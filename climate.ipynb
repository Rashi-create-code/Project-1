```python
import pandas as pd
```

```python
df = pd.read_csv("climate_change_dataset.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataframe is named df
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Year", y="CO2 Emissions (Tons/Capita)", hue="Country", marker="o")
plt.title("CO2 Emissions Over Time")
plt.ylabel("CO2 Emissions (Tons/Capita)")
plt.xlabel("Year")
plt.legend(title="Country")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="CO2 Emissions (Tons/Capita)", y="Avg Temperature (°C)", hue="Country", size="Population", sizes=(20, 200))
plt.title("Temperature vs CO2 Emissions")
plt.xlabel("CO2 Emissions (Tons/Capita)")
plt.ylabel("Avg Temperature (°C)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Country", y="Renewable Energy (%)", palette="viridis")
plt.title("Renewable Energy Usage by Country")
plt.ylabel("Renewable Energy (%)")
plt.show()
```
